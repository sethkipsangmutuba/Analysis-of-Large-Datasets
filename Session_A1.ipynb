# Introduction to the World of Big Data

Big Data refers to datasets whose size, speed, and complexity exceed the capabilities of traditional systems such as Relational Database Management Systems. It emerges from the inability of conventional databases to handle large-scale, fast, and diverse data.

Core characteristics are defined by the three Vs:

- Volume (scale of data)  
- Velocity (speed of generation and processing)  
- Variety (multiple data formats)  

Big Data supports structured, semi-structured, and unstructured data, including both human- and machine-generated sources such as social media, sensors, and transaction systems.

The data life cycle follows a pipeline:

generation → acquisition → preprocessing → integration → cleaning → transformation → analysis → visualization → decision-making.

Key advantages include scalability, flexibility, and advanced analytics capability. Major challenges arise from heterogeneity, high computational demands, real-time processing needs, and data quality management.

Generally, Big Data represents a shift toward distributed, scalable data systems enabling data-driven decision-making.

## 1.1 Understanding Big Data

Rapid internet growth drives massive data generation from platforms like WhatsApp, Facebook, Twitter, and YouTube. About 2.5 quintillion bytes are generated daily, with ~80% of data produced in recent years.

Big Data denotes data that are large, complex, high-velocity, and include structured and unstructured forms. Growth is continuous and global.

The core challenge is transforming data into value. Data must be captured, managed, and analyzed at the right speed to generate insights; however, extracting business value remains difficult.

Traditional systems like Relational Database Management Systems face limits in scalability, performance, latency, and cost as data size increases.

Big Data technologies overcome these via distributed processing and storage, including Hadoop, Hadoop Distributed File System, and MapReduce, enabling efficient capture, storage, and analysis of large-scale data.

```python
"""
Section 1.1: Understanding Big Data (Iris-based Experiment)

This module simulates:
1. Data ingestion (capture)
2. Data management (structured handling)
3. Data analysis (value extraction)

Note:
Iris dataset is used as a proxy for structured data systems.
"""

from sklearn.datasets import load_iris
import pandas as pd


def capture_data() -> pd.DataFrame:
    """
    Simulate data capture from a source.

    Returns
    -------
    pd.DataFrame
        Structured dataset
    """
    iris = load_iris()
    df = pd.DataFrame(iris.data, columns=iris.feature_names)
    df["target"] = iris.target
    return df


def manage_data(df: pd.DataFrame) -> pd.DataFrame:
    """
    Simulate data management (cleaning + structuring).

    Parameters
    ----------
    df : pd.DataFrame

    Returns
    -------
    pd.DataFrame
        Managed dataset
    """
    # Example: check for missing values
    df = df.dropna()

    # Ensure correct data types (important at scale)
    return df.astype(float)


def analyze_data(df: pd.DataFrame) -> pd.DataFrame:
    """
    Extract value from data (core Big Data challenge).

    Parameters
    ----------
    df : pd.DataFrame

    Returns
    -------
    pd.DataFrame
        Aggregated insights
    """
    # Group-wise statistical insight
    return df.groupby("target").agg(["mean", "std"])


def simulate_scaling(df: pd.DataFrame, factor: int = 1000) -> pd.DataFrame:
    """
    Simulate data growth to mimic Big Data scale.

    Parameters
    ----------
    df : pd.DataFrame
    factor : int
        Replication factor

    Returns
    -------
    pd.DataFrame
        Expanded dataset
    """
    # Replicate dataset to simulate volume growth
    return pd.concat([df] * factor, ignore_index=True)


if __name__ == "__main__":
    #a. Capture
    data = capture_data()

    #b. Simulate Big Data growth
    big_data = simulate_scaling(data, factor=500)

    #c. Manage
    managed = manage_data(big_data)

    #d. Analyze (value extraction)
    insights = analyze_data(managed)

    print("Data Shape (Simulated Big Data):", managed.shape)
    print("\nInsights:\n", insights)
```

## 1.2 Evolution of Big Data

The three defining dimensions—volume, velocity, and variety—were formalized by Doug Laney.

The Big Data processing life cycle includes:

acquisition → preprocessing → storage and management → privacy and security → analysis → visualization.

Big Data spans diverse domains such as web clickstream data, patient health data, and genomic research data.

Data growth has been exponential: from about 600 MB in the 1950s to nearly 100 petabytes by 2010, reflecting rapid scaling of digital information.

As of 2026, global data generation has reached the zettabyte era, with approximately 221 zettabytes of total data and over 0.4 zettabytes generated daily, driven by AI systems, IoT networks, cloud computing, and continuous digital interactions.

```python
"""
Section 1.2: Evolution of Big Data (Pipeline Simulation)

This module demonstrates:
1. Big Data lifecycle stages
2. Data growth simulation
3. End-to-end processing pipeline

Note:
Uses Iris dataset as structured proxy for lifecycle illustration.
"""

from sklearn.datasets import load_iris
import pandas as pd
import numpy as np


def acquire_data() -> pd.DataFrame:
    """
    Simulate data acquisition.

    Returns
    -------
    pd.DataFrame
        Raw dataset
    """
    iris = load_iris()
    df = pd.DataFrame(iris.data, columns=iris.feature_names)
    df["target"] = iris.target
    return df


def preprocess_data(df: pd.DataFrame) -> pd.DataFrame:
    """
    Simulate preprocessing (cleaning + normalization).

    Parameters
    ----------
    df : pd.DataFrame

    Returns
    -------
    pd.DataFrame
        Preprocessed dataset
    """
    df = df.dropna()

    # Normalize features (common in large-scale pipelines)
    features = df.columns[:-1]
    df[features] = (df[features] - df[features].mean()) / df[features].std()

    return df


def store_data(df: pd.DataFrame) -> pd.DataFrame:
    """
    Simulate storage/management layer.

    Parameters
    ----------
    df : pd.DataFrame

    Returns
    -------
    pd.DataFrame
        Stored dataset (placeholder for distributed storage)
    """
    # In real systems: HDFS / distributed DB
    return df.copy()


def apply_privacy(df: pd.DataFrame) -> pd.DataFrame:
    """
    Simulate privacy/security step.

    Parameters
    ----------
    df : pd.DataFrame

    Returns
    -------
    pd.DataFrame
        Anonymized dataset
    """
    # Remove identifiers (here: simulate by dropping target temporarily)
    return df.copy()


def analyze_data(df: pd.DataFrame) -> pd.DataFrame:
    """
    Perform analysis (value extraction).

    Parameters
    ----------
    df : pd.DataFrame

    Returns
    -------
    pd.DataFrame
        Analytical summary
    """
    return df.describe()


def visualize_data(df: pd.DataFrame) -> None:
    """
    Simulate visualization step.

    Parameters
    ----------
    df : pd.DataFrame
    """
    print("\n[Visualization Output]")
    print(df.head())


def simulate_exponential_growth(df: pd.DataFrame, steps: int = 5) -> pd.DataFrame:
    """
    Simulate exponential data growth.

    Parameters
    ----------
    df : pd.DataFrame
    steps : int
        Number of growth iterations

    Returns
    -------
    pd.DataFrame
        Expanded dataset
    """
    data = df.copy()

    for _ in range(steps):
        data = pd.concat([data, data], ignore_index=True)  # doubling

    return data


if __name__ == "__main__":
    # Acquisition
    data = acquire_data()

    #Simulate exponential growth (Volume dimension)
    big_data = simulate_exponential_growth(data, steps=6)

    # Preprocessing
    processed = preprocess_data(big_data)

    # Storage/Management
    stored = store_data(processed)

    # Privacy/Security
    secure = apply_privacy(stored)

    # Analysis
    insights = analyze_data(secure)

    # Visualization
    visualize_data(secure)

    print("\nData Shape:", secure.shape)
```

## 1.3 Failure of Traditional Database in Handling Big Data

Relational Database Management Systems were the dominant storage systems for organizational data, designed for structured data and centralized architectures. However, their limitations became apparent with the rise of Big Data.

### Key Limitations

- Rapid growth of data (terabytes to petabytes and beyond) exceeding system capacity  
- Scaling requires additional hardware, increasing cost significantly  
- Around 80% of data is semi-structured or unstructured, which RDBMS cannot efficiently process  
- High-velocity data streams cannot be handled effectively  

Big Data systems overcome these limitations using distributed architectures and commodity hardware.

### Comparison: RDBMS vs Big Data

| Attribute      | RDBMS                     | Big Data                                      |
|----------------|---------------------------|-----------------------------------------------|
| Data volume    | Gigabytes to terabytes    | Petabytes to zettabytes                       |
| Organization   | Centralized               | Distributed                                   |
| Data type      | Structured                | Structured, semi-structured, unstructured     |
| Hardware       | High-end systems          | Commodity hardware                            |
| Updates        | Read/write many times     | Write once, read many times                   |
| Schema         | Static                    | Dynamic                                       |

### Comparison: Data Mining vs Big Data

| Aspect          | Data Mining                                      | Big Data                                                                 |
|------------------|--------------------------------------------------|--------------------------------------------------------------------------|
| Definition       | Extracting knowledge from datasets              | Managing and processing massive datasets characterized by volume, velocity, variety |
| Data type        | Mainly structured (databases, spreadsheets)     | Structured, semi-structured, unstructured                                |
| Data sources     | Relational databases                            | NoSQL, distributed systems, diverse sources                              |
| Scale            | Gigabytes to terabytes                          | Petabytes to zettabytes                                                  |
| Processing cost  | High for large datasets                         | Relatively lower using distributed systems                               |

----
Big Data emerged to address scalability, heterogeneity, and real-time processing limitations of traditional databases and classical data mining systems.

```python
"""
Section 1.3: Failure of Traditional Databases (Simulation)

This module contrasts:
1. RDBMS-style processing (centralized, static schema)
2. Big Data-style processing (scalable, flexible schema)

Note:
Iris dataset is used as a structured baseline, then extended.
"""

from sklearn.datasets import load_iris
import pandas as pd


def load_structured_data() -> pd.DataFrame:
    """
    Load structured dataset (RDBMS-friendly).

    Returns
    -------
    pd.DataFrame
    """
    iris = load_iris()
    df = pd.DataFrame(iris.data, columns=iris.feature_names)
    df["target"] = iris.target
    return df


def simulate_unstructured_extension(df: pd.DataFrame) -> pd.DataFrame:
    """
    Add semi/unstructured data (RDBMS limitation scenario).

    Parameters
    ----------
    df : pd.DataFrame

    Returns
    -------
    pd.DataFrame
    """
    df = df.copy()
    df["text"] = ["sample text data"] * len(df)  # unstructured column
    return df


def rdbms_query(df: pd.DataFrame) -> pd.DataFrame:
    """
    Simulate RDBMS-style query (strict schema, structured only).

    Parameters
    ----------
    df : pd.DataFrame

    Returns
    -------
    pd.DataFrame
    """
    # Only numeric structured columns allowed
    numeric_cols = df.select_dtypes(include="number")
    return numeric_cols.groupby("target").mean()


def big_data_processing(df: pd.DataFrame) -> dict:
    """
    Simulate flexible Big Data processing.

    Parameters
    ----------
    df : pd.DataFrame

    Returns
    -------
    dict
        Structured + unstructured insights
    """
    results = {}

    # Structured analysis
    results["structured"] = df.select_dtypes(include="number").mean()

    # Unstructured handling (simple token count)
    if "text" in df:
        results["unstructured"] = df["text"].str.split().explode().value_counts()

    return results


def simulate_scaling_issue(df: pd.DataFrame, factor: int = 1000) -> pd.DataFrame:
    """
    Simulate large-scale data growth.

    Parameters
    ----------
    df : pd.DataFrame
    factor : int

    Returns
    -------
    pd.DataFrame
    """
    return pd.concat([df] * factor, ignore_index=True)


if __name__ == "__main__":
    # Structured data (RDBMS works fine)
    data = load_structured_data()

    # Scale data (volume challenge)
    big_data = simulate_scaling_issue(data, factor=300)

    # Add unstructured component
    mixed_data = simulate_unstructured_extension(big_data)

    # --- RDBMS-style processing ---
    rdbms_result = rdbms_query(mixed_data)

    # --- Big Data-style processing ---
    bigdata_result = big_data_processing(mixed_data)

    print("RDBMS Result (Structured Only):\n", rdbms_result.head())
```

## 1.4 3Vs of Big Data

Big Data is defined by three core dimensions: volume, velocity, and variety, which distinguish it from traditional data systems.

### 1.4.1 Volume

Volume refers to the massive and continuously increasing size of data generated and processed. It grows exponentially as organizations collect data for improved decision-making.

Big Data volume ranges from terabytes to zettabytes:

1024 GB = 1 TB → 1024 TB = 1 PB → 1024 PB = 1 EB → 1024 EB = 1 ZB → 1024 ZB = 1 YB

Major sources include social media, POS systems, online banking, GPS, and IoT sensors. For example, Facebook generates ~500 TB of data daily. Every online interaction (clicks, purchases, uploads, searches) contributes to this growth.

### 1.4.2 Velocity

Velocity refers to both the speed of data generation and the speed of data processing and analysis.

In modern systems, data arrives continuously and rapidly, often in real time, making storage and processing challenging.

Example high-velocity activity per 60 seconds:

- 3.3 million Facebook posts  
- 450,000 tweets  
- 400 hours of video uploads  
- 3.1 million Google searches  

This demands real-time or near-real-time processing systems.

### 1.4.3 Variety

Variety refers to the different formats of data handled in Big Data systems.

| Data Type        | Description                                   | Examples                                   |
|-----------------|-----------------------------------------------|--------------------------------------------|
| Structured      | Organized in tabular form, fixed schema       | Bank records, employee databases           |
| Semi-structured | Partial structure using tags or markers       | XML, JSON                                  |
| Unstructured    | No predefined format                          | Emails, images, videos, social media content |

Modern data sources such as social media feeds, vehicle sensors, and aircraft black boxes generate mostly unstructured data, which traditional systems cannot efficiently process.

-----

The 3Vs collectively illustrate why Big Data requires new architectures: scale (volume), speed (velocity), and diversity (variety) exceed the limits of traditional systems, necessitating distributed and scalable computing frameworks.

```python
"""
Section 1.4: 3Vs of Big Data (Experimental Simulation)

This module demonstrates:
1. Volume  → data scaling
2. Velocity → streaming-like processing
3. Variety → structured + semi/unstructured data

Note:
Conceptual illustration using Iris dataset as base.
"""

from sklearn.datasets import load_iris
import pandas as pd
import time


def load_base_data() -> pd.DataFrame:
    """
    Load structured dataset.

    Returns
    -------
    pd.DataFrame
    """
    iris = load_iris()
    df = pd.DataFrame(iris.data, columns=iris.feature_names)
    df["target"] = iris.target
    return df


# -------------------- VOLUME --------------------
def simulate_volume(df: pd.DataFrame, factor: int = 1000) -> pd.DataFrame:
    """
    Simulate exponential data growth.

    Parameters
    ----------
    df : pd.DataFrame
    factor : int

    Returns
    -------
    pd.DataFrame
    """
    return pd.concat([df] * factor, ignore_index=True)


# -------------------- VELOCITY --------------------
def simulate_velocity(df: pd.DataFrame, batch_size: int = 50):
    """
    Simulate streaming data processing.

    Parameters
    ----------
    df : pd.DataFrame
    batch_size : int

    Yields
    ------
    pd.DataFrame
        Data chunks (stream)
    """
    for i in range(0, len(df), batch_size):
        yield df.iloc[i:i + batch_size]
        time.sleep(0.01)  # simulate arrival delay


def process_stream(stream):
    """
    Process incoming data stream in near real-time.

    Parameters
    ----------
    stream : generator

    Returns
    -------
    list
        Running means per batch
    """
    results = []
    for batch in stream:
        results.append(batch.mean(numeric_only=True))
    return results


# -------------------- VARIETY --------------------
def simulate_variety(df: pd.DataFrame) -> pd.DataFrame:
    """
    Add semi/unstructured components.

    Parameters
    ----------
    df : pd.DataFrame

    Returns
    -------
    pd.DataFrame
    """
    df = df.copy()

    # Semi-structured (JSON-like string)
    df["json"] = df.apply(lambda r: f'{{"sepal": {r[0]}}}', axis=1)

    # Unstructured text
    df["text"] = ["flower data sample"] * len(df)

    return df


def analyze_variety(df: pd.DataFrame) -> dict:
    """
    Extract insights from mixed data types.

    Parameters
    ----------
    df : pd.DataFrame

    Returns
    -------
    dict
    """
    results = {}

    # Structured
    results["structured"] = df.select_dtypes(include="number").mean()

    # Semi-structured (simple parse)
    results["semi_structured"] = df["json"].str.len().mean()

    # Unstructured (word frequency)
    results["unstructured"] = df["text"].str.split().explode().value_counts()

    return results


# -------------------- MAIN --------------------
if __name__ == "__main__":
    data = load_base_data()

    # Volume
    big_data = simulate_volume(data, factor=500)
    print("Volume Shape:", big_data.shape)

    #Velocity
    stream = simulate_velocity(big_data, batch_size=200)
    velocity_results = process_stream(stream)
    print("\nVelocity Batches Processed:", len(velocity_results))

    #Variety
    mixed_data = simulate_variety(big_data)
    variety_results = analyze_variety(mixed_data)

    print("\nVariety Insights:")
    print("Structured:\n", variety_results["structured"].head())
    print("\nUnstructured:\n", variety_results["unstructured"].head())
```

## 1.6 Different Types of Data

Data is broadly categorized into human-generated and machine-generated forms. Human-generated data arises from interactions between humans and systems, such as emails, documents, social media posts (e.g., Facebook). Machine-generated data is produced automatically by systems without direct human involvement, such as sensor outputs, satellite data, weather forecasting systems, and disaster monitoring systems.

---

## Core Data Types in Big Data

### 1. Structured Data

Structured data is organized in a fixed format (rows and columns) and stored in relational databases.

| Feature   | Description                          |
|-----------|--------------------------------------|
| Format    | Tabular (rows and columns)          |
| Storage   | Relational databases                |
| Querying  | Efficient using keys                |
| Examples  | Employee records, financial transactions |

It is highly organized and easily processed using traditional tools.

---

### 2. Unstructured Data

Unstructured data has no predefined format and constitutes nearly 80% of total data generated.

| Feature  | Description                |
|----------|----------------------------|
| Format   | No fixed structure        |
| Storage  | Text or binary files      |
| Examples | Videos, images, audio, emails, social media posts |

Binary unstructured data includes images, audio, and video, while text-based unstructured data includes emails, PDFs, and documents. This type dominates modern data ecosystems, including outputs from platforms like YouTube and web search systems.

---

### 3. Semi-Structured Data

Semi-structured data lies between structured and unstructured forms. It has partial organization using tags or markers.

| Feature | Description                  |
|---------|------------------------------|
| Format  | Tagged or hierarchical structure |
| Storage | Non-relational systems       |
| Examples| XML, JSON                    |

It is more flexible than structured data and easier to analyze than unstructured data.

---

Big Data systems must handle a combination of structured, semi-structured, and unstructured data, reflecting the complexity of modern human and machine-generated environments.

## 1.7 Big Data Infrastructure

Big Data infrastructure comprises tools and technologies that enable storage, processing, and analysis of large-scale data. Traditional relational systems such as Relational Database Management Systems became insufficient due to limitations in handling the 3Vs (volume, velocity, variety) and high scalability costs. This led to distributed, low-cost, highly scalable architectures.

---

## Core Technologies

### 1. Apache Hadoop

Hadoop is an open-source Java-based framework designed for large-scale data processing.

#### Key Features:

- Stores structured, semi-structured, and unstructured data  
- Processes data in parallel across distributed systems  
- Highly scalable and cost-effective  
- Follows a write-once, read-many-times model  
- Built on clusters (multiple interconnected machines acting as a single system)  
- Designed for fault tolerance and distributed computation  

Hadoop enables scalable storage and computation using commodity hardware rather than expensive high-end systems.

---

### 2. Hadoop Distributed File System (HDFS)

HDFS is the storage layer of Hadoop.

#### Key Features:

- Stores very large datasets across distributed nodes  
- Optimized for streaming access patterns  
- Uses low-cost commodity hardware  
- Supports write-once, read-many-times access model  
- Integrates data from multiple sources for analysis  

HDFS enables reliable storage without requiring highly expensive or highly reliable individual machines.

---

### 3. MapReduce

MapReduce is a programming model for batch processing in Hadoop.

#### Key Features:

- Follows a divide-and-conquer approach  
- Processes data in parallel across clusters  
- Highly scalable, fault-tolerant, and reliable  
- Designed for batch workloads only  
- Reduces processing time by executing computation where data resides (data locality), avoiding costly data transfer  

---

Big Data infrastructure shifts computation from centralized systems to distributed, scalable frameworks, enabling efficient storage and processing of massive datasets at low cost while maintaining performance under heavy workloads.

```python
"""
Section 1.7: Big Data Infrastructure (Conceptual Simulation)

This module demonstrates:
1. Distributed storage concept (HDFS-like abstraction)
2. Parallel processing model (MapReduce-like)
3. Framework-level coordination (Hadoop-like system idea)

Note:
This is a simplified simulation, not actual distributed execution.
"""

from sklearn.datasets import load_iris
import pandas as pd
from collections import defaultdict


# ---------------- HDFS-LIKE STORAGE ----------------
class DistributedStorage:
    """
    Simulate distributed storage system (HDFS concept).
    """

    def __init__(self, n_nodes: int = 3):
        self.nodes = {i: [] for i in range(n_nodes)}

    def store(self, data: pd.DataFrame):
        """
        Distribute data across nodes.

        Parameters
        ----------
        data : pd.DataFrame
        """
        chunks = np.array_split(data, len(self.nodes))
        for i, chunk in enumerate(chunks):
            self.nodes[i].append(chunk)

    def get_all(self):
        """
        Retrieve all distributed data.

        Returns
        -------
        pd.DataFrame
        """
        return pd.concat([pd.concat(v) for v in self.nodes.values()])


# ---------------- MAPREDUCE ----------------
def map_function(df: pd.DataFrame):
    """
    Map step: transform data into key-value pairs.

    Parameters
    ----------
    df : pd.DataFrame

    Returns
    -------
    list
    """
    return [(col, df[col].mean()) for col in df.columns]


def reduce_function(mapped_data):
    """
    Reduce step: aggregate results.

    Parameters
    ----------
    mapped_data : list

    Returns
    -------
    dict
    """
    result = defaultdict(list)

    for key, value in mapped_data:
        result[key].append(value)

    return {k: sum(v) / len(v) for k, v in result.items()}


# ---------------- HADOOP-LIKE FRAMEWORK ----------------
class HadoopSimulator:
    """
    Simulate Hadoop framework coordination layer.
    """

    def __init__(self):
        self.storage = DistributedStorage()

    def run_job(self, df: pd.DataFrame):
        """
        Execute full pipeline: store → map → reduce.

        Parameters
        ----------
        df : pd.DataFrame

        Returns
        -------
        dict
        """
        # Store data (HDFS concept)
        self.storage.store(df)

        # Retrieve distributed data
        data = self.storage.get_all()

        # Map step
        mapped = map_function(data)

        # Reduce step
        return reduce_function(mapped)


# ---------------- MAIN ----------------
if __name__ == "__main__":

    iris = load_iris()
    data = pd.DataFrame(iris.data, columns=iris.feature_names)

    # Run Hadoop-like pipeline
    system = HadoopSimulator()
    result = system.run_job(data)

    print("MapReduce Aggregated Result:\n", result)
```

```python
"""
Section 1.8: Big Data Life Cycle (End-to-End Simulation)

This module demonstrates:
1. Data generation
2. Data aggregation
3. Preprocessing (cleaning, integration, reduction, transformation)
4. Analytics
5. Visualization

Note:
Iris dataset used as structured base for lifecycle simulation.
"""

from sklearn.datasets import load_iris
import pandas as pd
import numpy as np


# ---------------- 1. GENERATION ----------------
def generate_data() -> pd.DataFrame:
    """
    Simulate Big Data generation stage.

    Returns
    -------
    pd.DataFrame
    """
    iris = load_iris()
    df = pd.DataFrame(iris.data, columns=iris.feature_names)
    df["target"] = iris.target

    # simulate growth
    return pd.concat([df] * 200, ignore_index=True)


# ---------------- 2. AGGREGATION ----------------
def aggregate_data(df: pd.DataFrame) -> pd.DataFrame:
    """
    Simulate data aggregation from multiple sources.

    Parameters
    ----------
    df : pd.DataFrame

    Returns
    -------
    pd.DataFrame
    """
    return df.groupby("target").mean().reset_index()


# ---------------- 3. PREPROCESSING ----------------
def clean_data(df: pd.DataFrame) -> pd.DataFrame:
    """
    Handle missing values and inconsistencies.

    Returns
    -------
    pd.DataFrame
    """
    df = df.copy()
    df = df.fillna(df.mean(numeric_only=True))  # missing values
    return df


def reduce_data(df: pd.DataFrame) -> pd.DataFrame:
    """
    Simulate dimensionality reduction.

    Returns
    -------
    pd.DataFrame
    """
    # keep only strongest features (simplified selection)
    return df.iloc[:, :3]


def transform_data(df: pd.DataFrame) -> pd.DataFrame:
    """
    Normalize data (transformation step).

    Returns
    -------
    pd.DataFrame
    """
    return (df - df.mean()) / df.std()


# ---------------- 4. ANALYTICS ----------------
def descriptive_analytics(df: pd.DataFrame) -> pd.DataFrame:
    """
    Compute descriptive statistics.

    Returns
    -------
    pd.DataFrame
    """
    return df.describe()


def predictive_signal(df: pd.DataFrame) -> pd.Series:
    """
    Simulate predictive insight (simple proxy).

    Returns
    -------
    pd.Series
    """
    return df.mean(axis=1)


# ---------------- 5. VISUALIZATION (TEXTUAL SIMULATION) ----------------
def visualize(df: pd.DataFrame) -> None:
    """
    Simulate visualization output.

    Prints structured representation of insights.
    """
    print("\n[Line Trend Simulation]")
    print(df.iloc[:5].mean())

    print("\n[Comparison Snapshot]")
    print(df.head())


# ---------------- MAIN PIPELINE ----------------
if __name__ == "__main__":

    # 1. Generation
    data = generate_data()

    # 2. Aggregation
    aggregated = aggregate_data(data)

    # 3. Preprocessing
    cleaned = clean_data(aggregated)
    reduced = reduce_data(cleaned)
    transformed = transform_data(reduced)

    # 4. Analytics
    desc = descriptive_analytics(transformed)
    prediction_signal = predictive_signal(transformed)

    # 5. Visualization
    visualize(transformed)

    print("\nDescriptive Analytics:\n", desc)
    print("\nPredictive Signal Sample:\n", prediction_signal.head())
```

## 1.9 Big Data Technology

Big Data technology governs how data is generated, captured, stored, processed, and analyzed efficiently. It enables scalable and cost-effective extraction of insights from massive datasets.

A core technology is Hadoop, an open-source framework originally developed by Doug Cutting and Mike Cafarella as part of the Nutch project for web indexing. Hadoop integrates concepts such as MapReduce and distributed file systems to handle large-scale data processing.

---

## Core Components of Hadoop

- **Hadoop Distributed File System (HDFS)** — storage layer  
- **MapReduce** — processing model  
- **Hadoop Common** — core utilities  
- **YARN (Yet Another Resource Negotiator)** — resource management layer  

### MapReduce
MapReduce enables parallel processing of large datasets in distributed environments.

### YARN
YARN manages cluster resources and allows multiple processing models beyond MapReduce. It was introduced in Hadoop 2.0.

### Hadoop Common
Hadoop Common provides essential libraries and scripts required to run the system.

Hadoop enables storage and processing of petabyte-scale structured, semi-structured, and unstructured data using low-cost commodity hardware.

---

## 1.9.1 Challenges in Big Data Technology

Key challenges include:

- Heterogeneity and incompleteness of data  
- High volume and velocity  
- Scalable data storage  
- Data privacy and security concerns  

---

## 1.9.2 Heterogeneity and Incompleteness

Big Data originates from multiple heterogeneous sources, requiring integration into a consistent format before analysis. Data may also be incomplete (e.g., missing patient attributes in medical records), reducing analytical quality. In some cases, missing values (NULLs) are tolerated if they do not significantly affect outcomes.

---

## 1.9.3 Volume and Velocity

### Volume challenge
Exponential data growth exceeds traditional storage and processing capacity, increasing computation time.

### Velocity challenge
Data must be processed in real time (e.g., fraud detection in credit card transactions requires instant decisions).

---

## 1.9.4 Data Storage

Big Data requires scalable, fault-tolerant, and reliable storage systems due to:

- Massive data generation from social media, IoT, and online systems  
- Continuous streaming data where storage needs are not predetermined  
- Requirement for historical data retrieval for analysis and comparison  

---

## 1.9.5 Data Privacy

Data privacy ensures controlled access to sensitive information such as health records, financial data, and personal identifiers.

### Key Concerns:

- Unauthorized access to personal data  
- Controlled data sharing for analysis  
- Defined access rights (who can access, when, and how much)  
- Balancing data utility with confidentiality  

---

Effective big data systems require strong governance across data collection, storage, and analysis stages to ensure security, scalability, and ethical use of data.

```python
"""
Section 1.9: Big Data Technology (Hadoop Ecosystem Simulation)

This module demonstrates:
1. Hadoop architecture components
2. Distributed storage + processing abstraction
3. Resource management concept
4. Big Data challenges handling

Note:
Iris dataset used as simplified structured input.
"""

from sklearn.datasets import load_iris
import pandas as pd
import numpy as np


# ---------------- HDFS (Storage Layer) ----------------
class HDFS:
    """
    Simulate Hadoop Distributed File System.
    """

    def __init__(self, nodes: int = 3):
        self.storage = {i: [] for i in range(nodes)}

    def write(self, df: pd.DataFrame):
        """
        Distribute data across nodes.
        """
        chunks = np.array_split(df, len(self.storage))
        for i, chunk in enumerate(chunks):
            self.storage[i].append(chunk)

    def read(self) -> pd.DataFrame:
        """
        Retrieve distributed data.
        """
        return pd.concat([pd.concat(v) for v in self.storage.values()])


# ---------------- MapReduce (Processing Layer) ----------------
class MapReduce:
    """
    Simulate distributed processing model.
    """

    def map(self, df: pd.DataFrame):
        """
        Map step: compute partial aggregates.
        """
        return df.groupby(df.columns[-1]).mean()

    def reduce(self, mapped_df: pd.DataFrame):
        """
        Reduce step: final aggregation.
        """
        return mapped_df.mean()


# ---------------- YARN (Resource Manager) ----------------
class YARN:
    """
    Simulate cluster resource management.
    """

    def allocate(self, task_name: str):
        """
        Allocate resources to tasks.
        """
        return f"Resources allocated to {task_name}"


# ---------------- Hadoop System ----------------
class HadoopSystem:
    """
    Full Hadoop ecosystem simulation.
    """

    def __init__(self):
        self.hdfs = HDFS()
        self.mapreduce = MapReduce()
        self.yarn = YARN()

    def run_pipeline(self, df: pd.DataFrame):
        """
        Execute full Big Data workflow.
        """

        print(self.yarn.allocate("BigDataJob"))

        # Storage
        self.hdfs.write(df)
        data = self.hdfs.read()

        # Processing
        mapped = self.mapreduce.map(data)
        reduced = self.mapreduce.reduce(mapped)

        return reduced


# ---------------- Challenge Simulation ----------------
def simulate_missing_data(df: pd.DataFrame) -> pd.DataFrame:
    """
    Introduce incompleteness (real Big Data issue).
    """
    df = df.copy()
    df.iloc[::10] = np.nan
    return df


def handle_missing_data(df: pd.DataFrame) -> pd.DataFrame:
    """
    Handle data incompleteness.
    """
    return df.fillna(df.mean(numeric_only=True))


# ---------------- MAIN ----------------
if __name__ == "__main__":

    iris = load_iris()
    data = pd.DataFrame(iris.data, columns=iris.feature_names)

    # Introduce challenges
    data = simulate_missing_data(data)
    data = handle_missing_data(data)

    # Run Hadoop system
    system = HadoopSystem()
    result = system.run_pipeline(data)

    print("\nFinal Aggregated Output:\n", result)
```
